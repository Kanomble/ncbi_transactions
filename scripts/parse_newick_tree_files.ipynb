```python
import pandas as pd
import re
```

```python
r_dataframe = '../data/curvibacter_rec_blasts/eps_operon_vs_all_complete_chromosome/result_dataframe.csv'
result_df = pd.read_csv(r_dataframe,header=0,index_col=0)
```

```python
rf = result_df[result_df['sscinames'] == 'Rhodoferax sp. AJA081-3']
```

```python
import os
```

```python
folder = os.listdir('../data/curvibacter_eps_operon_17_organisms/rec_blast_curvi_processing')
```

```python
files = []
outfiles = []
for path in folder:
    if 'WP' in path:
        n = '../data/curvibacter_eps_operon_17_organisms/rec_blast_curvi_processing/'+path+'/tree.nwk'
        m = '../data/curvibacter_eps_operon_17_organisms/rec_blast_curvi_processing/'+path+'/tree_annot_with_genus.nwk'
        files.append(n)
        outfiles.append(m)
```

```python
for infile, outfile in zip(files,outfiles):
    regex = re.compile('WP_[0-9]{9}')
    treefile = open(infile,'r')
    lines = treefile.readlines()
    treefile.close()

    newick = lines[0]
    wps = re.findall(regex,lines[0])
    for wp in wps:
        name = list(result_df[result_df['sacc'] == wp]['genus'])[0]
        print(name)
        newick = re.sub(wp,name,newick)

    output = open(outfile,'w')
    output.write(newick)
    output.close()
```
