# Working with databases

```python
#available search tags
!einfo -db assembly -fields
```

```python
#available format specifiers
!esearch -db assembly -query "Curvibacter[ORGN] AND (contig[ASLV] OR chromosome[ASLV])" | efetch -format docsum
```

```python
!esearch -db assembly -query "Curvibacter[ORGN] AND (contig[ASLV] OR chromosome[ASLV])" | efetch -format docsum | xtract -pattern DocumentSummary -element AssemblyName AssemblyStatus
```

```python
!esearch -db assembly -query "Curvibacter [ORGN]" | efetch -format docsum | xtract -pattern DocumentSummary -element Id AssemblyAccession AssemblyName AssemblyStatus SubmissionDate FtpPath_GenBank Taxid Organism SpeciesTaxid SpeciesName > ../data/entrez/curvibacter_assemblies.table
```

```python
cols = ['Id', 'AssemblyAccession', 'AssemblyName', 'AssemblyStatus', 'SubmissionDate','FtpPath_GenBank', 'Taxid', 'Organism', 'SpeciesTaxid', 'SpeciesName']
assemblies = pd.read_table('../data/entrez/curvibacter_assemblies.table',header=None)

assemblies.columns = cols
assemblies.head()
```
