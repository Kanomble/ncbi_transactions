```python
def change_alignment_header(original_msa_file,new_msa_file):
    try:
        print("[*] START changing headers of original alignment file")
        with open(original_msa_file,'r') as infile:
            with open(new_msa_file,'w') as newfile:
                for line in infile.readlines():
                    if line.startswith('>'):
                        newfile.write(line.split(' ')[0]+'\n')
                    else:
                        newfile.write(line)
    except Exception as e:
        print("[-] ERROR during changing header of alignment file ...")
        raise Exception("[-] ERROR with exception: {}".format(e))
    print("[+] DONE")        
    return 0
```
