```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
transport_file = '../data/lps_synthesis/lps_transport.csv'
lipidA_file = '../data/lps_synthesis/lipidA_synthesis.csv'
core_oligo_file = '../data/lps_synthesis/core_oligosaccharides.csv'
o_antigen_file = '../data/lps_synthesis/o_antigen_synthesis.csv'
wzy_file = '../data/lps_synthesis/lps_synthesis_ligase.csv'
```

```python
transport = pd.read_csv(transport_file,index_col=0,header=0)
```

```python
def get_plot_data(dataframe):
    dataframes = []
    for qseqid in dataframe['qseqid'].unique():
        dataframes.append(dataframe[dataframe['qseqid'] == qseqid])
    
    ydata = []
    xticklabels = []
    for df in dataframes:
        counter = 0
        transient_list = []
        for ss in df['sseqid'].str.split("_"):
            if ss[0] not in transient_list:
                transient_list.append(ss[0])
                counter += 1
        ydata.append(counter)
        xticklabels.append(df['stitle'].to_list()[0])
    return ydata, xticklabels
```

```python
ydata, yticklabels = get_plot_data(transport)
plt.figure(figsize=(8,6))
yticklabels = ["LptA","LptB","LptC","LptD","LptE","LptF","LptG","MsbA"]
bar = plt.bar(height=ydata,x=range(1,len(ydata)+1),color='royalblue',edgecolor='black',linewidth=2)
plt.ylim(0,40)
plt.xlabel("LPS Transport Genes",fontsize=15)
#plt.ylabel("Orthologs in different $\it{Curvibacter}$ assemblies", fontsize=15)
plt.xticks(ticks=range(1,len(ydata)+1),labels=yticklabels, rotation=90, fontsize=15)
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height:.0f}', ha='center', va='bottom', fontsize=15)
plt.tight_layout()
plt.savefig("../data/lps_synthesis/transport_without_text.png",dpi=400)
```
